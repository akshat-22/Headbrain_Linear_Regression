###  Now dataset is clean as it not contain any NULL value or NAN(not a number) value

```python
data.head(2)
```

```python
sex=pd.get_dummies(data['Sex'],drop_first=True)
sex.head(5)
```

```python
embark=pd.get_dummies(data["Embarked"],drop_first=True)
embark.head(5)
```

```python
Pcl=pd.get_dummies(data["Pclass"],drop_first=True)
Pcl.head(5)
```

```python
data=pd.concat([data,sex,embark,Pcl],axis=1)
```

```python
data.head(5)
```

```python
data.drop(['Pclass','Embarked','Sex','Name','Ticket','PassengerId'], axis=1,inplace=True )
```

```python
data.head(5)
```

### Train Data

```python
X=data.drop(["Survived"],axis=1) #Independent Variable
y=data["Survived"]              #Dependent Variable
```

```python
X.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```
